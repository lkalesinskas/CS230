```python
from nilearn import image
import numpy as np
import os

#import files
path = "E:\\CS230\\"
metadataFiles = []
anatomicalMRI = []
restingStateFMRI = []
for dirpath, subdirs, files in os.walk(path):
    for x in files:
        if x.endswith(".csv"):
            metadataFiles.append(os.path.join(dirpath, x))
        if x.endswith("anat.nii.gz"):
            anatomicalMRI.append(os.path.join(dirpath, x))
        if x.endswith("rest.nii.gz"):
            restingStateFMRI.append(os.path.join(dirpath, x))
        print(x)

print(metadataFiles)
print(anatomicalMRI[0:10])
print(restingStateFMRI[0:10])
```

```python
#metadata handling

healthy = {}
for fileName in metadataFiles:
    f = open(fileName, "r")
    f.readline()
    print(fileName)
    for line in f:
        splitted = line.strip().split(",")
        healthy[splitted[1]] = splitted[3]
    f.close()


f = open("C:\\Users\\Larry\\eclipse-workspace\\StupidFiles\\not_working_files.txt", "r")
for line in f:
    splitted = line.strip().split(",")
    healthy[splitted[0]] = splitted[1]
f.close()
print(healthy)
```

```python
def reshape(img):
    return image.resample_img(img, target_shape=(128, 128, 128), target_affine=np.eye(4))

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

count = 0
for i in anatomicalMRI:
    splitted = i.split("\\")
    id = ""
    healthyNumber = ""
    for string in splitted:
        if(is_number(string)):
            id = string
            healthyNumber = healthy[string]
    filename = i
    img = image.smooth_img(i, fwhm=None)
    print(img.shape)
    reshaped = reshape(img)
    print(reshaped.shape)
    if (healthyNumber == "1"):
        healthyNumber = "autism"
    elif (healthyNumber == "2"):
        healthyNumber = "healthy"
    else:
        healthyNumber = "unknown"
    outputfile = "C:\\Users\\Larry\\NilearnStuff\\FinalDataset\\" + id + "_" + healthyNumber + "_" + str(count) + "_mri.nii.gz"
    reshaped.to_filename(outputfile)
    print(outputfile)
    count +=1 
```

```python
def reshape(img):
    return image.resample_img(img, target_shape=(64, 64, 31), target_affine=np.eye(4))

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

count = 0
for i in restingStateFMRI:
    if count > 914:
        splitted = i.split("\\")
        id = ""
        healthyNumber = ""
        for string in splitted:
            if(is_number(string)):
                id = string
                healthyNumber = healthy[string]
        filename = i
        img = image.smooth_img(i, fwhm=None)
        print(img.shape)
        imageList = []
        reshaped = reshape(img)
        print(reshaped.shape)
        if (reshaped.shape[3]>=85):
            for a in range (0, 85):
                imageList.append(image.index_img(reshaped, a))
            joint_fMRI = image.concat_imgs(imageList)
            print(joint_fMRI.shape)
            if (healthyNumber == "1"):
                healthyNumber = "autism"
            elif (healthyNumber == "2"):
                healthyNumber = "healthy"
            else:
                healthyNumber = "unknown"
            outputfile = "C:\\Users\\Larry\\NilearnStuff\\FinalDataset\\" + id + "_" + healthyNumber + "_" + str(count) + "_fmri.nii.gz"
            joint_fMRI.to_filename(outputfile)
            print(outputfile)
    count +=1
```
